```python
def greedy_search(list,x):
  # for-loop to iterate through list
  for i in range(len(list)):
    if list[i] == x:
      statement = "Index {index} on list {list_arg} is {element} equal {value}. Found!".format(index = i,list_arg = list,element = list[i],value = x)
      print(statement)
      return i
    else:
      statement = "Index {index} on list {list_arg} is {element} not {value}. Move on!".format(index = i,list_arg = list,element = list[i],value = x)
      print(statement)

  # exiting/finishing loop means that value (x) was not found in list
  return -1
```

```python
def binary_search(list,x):
  start_ind = 0
  end_ind= len(list) - 1

  while start_ind<=end_ind:
    mid_ind = (start_ind + end_ind)//2
    if list[mid_ind] == x:
      statement = "Index {index} on list {list_arg} is {element} equal {value}. Found!".format(index = mid_ind,list_arg = list,element = list[mid_ind],value = x)
      print(statement)
      return mid_ind
    else:
      if list[mid_ind] < x:
        statement = "Index {index} on list {list_arg} is {element} less than {value}. Move up!".format(index = mid_ind,list_arg = list,element = list[mid_ind],value = x)
        start_ind = mid_ind + 1
      else:
        statement = "Index {index} on list {list_arg} is {element} greater than {value}. Move down!".format(index = mid_ind,list_arg = list,element = list[mid_ind],value = x)
        end_ind = mid_ind - 1
      
      print(statement)

  # exiting/finishing loop means that value (x) was not found in list
  return -1
```

## Testing the functions

```python
set1 = [1, 3, 5, 7, 9, 11, 12, 14, 16, 17, 21, 33, 45, 99, 101]
set2 = [1, 3, 5, 7, 9, 17, 12, 14, 16, 11, 21, 33, 45, 99, 101]
```

```python
# test greedy_search with set1
greedy_search(set1,17)
```

```python
# test binary_search with set1
binary_search(set1,17)
```

```python
# test greedy_search with set2
greedy_search(set2,17)
```

```python
# test binary_search with set2
binary_search(set2,17)
```
