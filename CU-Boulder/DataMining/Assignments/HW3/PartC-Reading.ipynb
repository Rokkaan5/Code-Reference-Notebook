**Paper title:** *Perceived usefulness of online customer reviews: A review mining approach using machine learning & exploratory data analysis*

# Question 1

**What problem is addressed? Why is the problem important and challenging?**

The paper is trying to answer what factors in a customer review determine how helpful the review is (in making a decision to purchase a product). 

The importance of the problem is that reviews can help consumers decide whether to purchase a product or not. In the past, online shoppers faced the problem of not knowing enough about a product to make a decision to go through with a purchase. So originally, the element of product reviews helped other consumers make a more informed decision. However, nowadays, there can be so many reviews that other consumers suffer from information overload and have difficulty determining which reviews are actually helpful. 

The challenge this particular paper faces is that there are a lot of studies with contradicting conclusions as to which factors correlate to how helpful a review is. 

# Question 2

**A high-level, brief description of the proposed solution (no need to include the details).**

The way that this paper is trying to address the problem of finding factor importance by using a hybrid model methodology.

The insights that the paper finds in the study are as follows. In "search products," they found that people rely more on star-rating rather than review content. Products that the paper calls "experience products" (like games) put more importance on the review content.
