```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
fatal = pd.read_csv('Fatalities.csv')
```

```python
fatal.head()
```

```python
fatal.drop(columns = ['Unnamed: 0'],inplace=True)
fatal.head()
```

```python
fatal.info()
```

```python
fatal.describe()
```

```python
plt.figure(figsize=(10,5))
sns.relplot(data=fatal,x='year',y='fatal',hue='state',kind='line')
```

```python
plt.plot(fatal['youngdrivers'],fatal['fatal'],'.')
plt.xlabel('# of young drivers')
plt.ylabel('Vehicle fatalities')
```

```python
sns.boxplot(data=fatal,x='drinkage',y='afatal')
plt.xlabel('Minimum Legal Drinking Age')
plt.ylabel('Alcohol-related Vehicle fatalities')
plt.xticks(rotation=90)
```

```python
plt.plot(fatal['beertax'],fatal['afatal'],'.')
plt.xlabel('State Tax on Case of Beer')
plt.ylabel('Alcohol-related Vehicle fatalities')
```

```python
sns.violinplot(data=fatal,x='breath',y='afatal')
plt.xlabel('Preliminary Breath Test Law In-Place')
plt.ylabel('Alcohol-related Vehicle fatalities')
```

```python
plt.plot(fatal['unemp'],fatal['fatal'],'.')
plt.xlabel('Unemployment Rate')
plt.ylabel('Vehicle fatalities')
```

```python
plt.bar(fatal['drinkage'],fatal['beertax'])
plt.xlabel('Minimum Legal Drinking Age')
plt.ylabel('State Tax on Case of Beer')
```

```python
sns.boxplot(data=fatal,x='drinkage',y='beertax')
plt.xticks(rotation=90)
plt.xlabel('Minimum Legal Drinking Age')
plt.ylabel('State Tax on Case of Beer')
```

```python
sns.boxplot(data=fatal,x='drinkage',y='youngdrivers')
plt.xticks(rotation=90)
plt.xlabel('Minimum Legal Drinking Age')
plt.ylabel('Percent of Young Drivers')
```
