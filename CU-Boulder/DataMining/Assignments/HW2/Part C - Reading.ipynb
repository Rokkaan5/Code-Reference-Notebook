# Part C - Reading

> Paper title: Exploratory Data Analysis 

## Question 1

What problem is addressed? Why is the problem important and challenging?

I don't know if I would say the paper outlines a specific "problem," but it is obvious that the overall paper is trying to stress the underrated importance of exploratory data analysis. The paper specifically reiterates the points made by John W. Tukey, who strongly advocated for data analysis to be considered a field of science in itself. 

I suppose some of the issues proposed by the paper are the concerns that come from underappreciating the power of exploratory data analysis, including not asking the right questions, not taking into consideration the potential flaws in the data collection process (thus the quality of the data), and not noticing patterns, outliers or relationships found in the data through data visualization. 

## Question 2

A high-level, brief description of the proposed solution (no need to include the details).

Again, the entire paper is centered around the topic of EDA, so the general "proposed solution" is (the tools in) EDA.

Visualization for inspection and communication. Robust and resistant measures for measures of spread. Etc. 

## Question 3

Discuss one point that you agree most with the author in detail. What is it, and why do you agree with the author?

The point that I probably agree with the author the most is the main point of the paper, exploratory data analysis can be and is a very powerful tool. 

The numerous times I have been amazed by the power of data visualization and EDA is probably one of my main inspirations in pursuing a program in data science.  Data visualization was an easy and quick way to demonstrate and illustrate the problem points that I wanted to emphasize in the introduction of my research projects. And then, as the paper mentioned, it was also useful for communicating my model results of the data. (To evaluate the accuracy of and analyze the model I built, etc.)

There's so much that EDA can tell us about our data (and models). I also like how the paper mentioned the use of EDA for further confirmation because I did the same thing to evaluate the quality of data I was working with by comparing the relationships and patterns to common physical theories during my work as an undergraduate research assistant in physics research.
