# Question 1

**What is the step Data Understanding in Statistics - in the Data Mining Pipeline? Why do we need it?**

I believe "Data Understanding" would still be part of the "Data Preprocessing" part of the Data Mining pipeline. 

The importance of the process almost seems pretty self-explanatory: doing anything with a set of data (data mining or otherwise) is much more meaningful if one actually understands the nature of the data they are working with. But to be more specific, data understanding includes evaluating statistical measures like central tendency, dispersion, and similarity for the quantitative values, as well as analyzing the types of qualitative data (categorical, etc.) within the dataset. 

Understanding all of this information in our data helps us know where to start when evaluating the data and determine what we can potentially do with it.  

# Question 2

**What are the measures of central tendency (Ms)? Explain the relevance of each of them with suitable examples.**

Measures of central tendency: Mean, median, and mode.

Mean is also called the average, and is a measure of the "center" of values based on maximum and minimum values and the total number of values. While it can have some tendency to be skewed by extreme maximum or minimum values, it probably also provides a nice balance between median and mode.

The median of data gives the actual "center" value if one lists all the available numerical values from smallest to largest and retains all the repeated occurrences. 

The mode is the numerical value in a set of data that occurs the most. 

Evaluating these values on their own can sometimes provide very little information and can potentially be misleading. But comparing all of them with each other can provide some more information. 

For example, suppose for a given set of data all these measures have the exact same value (or at least differ by very little). In that case, we can probably infer that the data stabilizes or "centralizes" around that value based on the meaning of each of them. 

But if they all differ from each other, we can probably infer that there's a great spread in the range of values and will probably gain more insight from other methods like dispersion. (But of course, this can also depend on how much data we actually have.)

# Question 3

**What are the measures of dispersion? Explain the relevance of each of them with suitable examples.**

Measures of dispersion: Range, quartiles, IQR, variance, and standard deviation. Much like the measures of central tendency gave an idea where a set of numerical data "centers" to, measures of dispersion give an idea of how the range of values is spread (or "dispersed") across the data. 

Range is the difference between the absolute maximum and minimum, thus (as the name suggests) gives an idea of how large the range of values is.

Quartiles are useful to provide information on multiple parts of the entire range of values in a set of data. The second quartile, or 50th percentile, is the same as the median. So then the first and third quantiles (25th and 75th percentile respectively) are the values halfway between the median and minimum and maximum values respectively. Knowing these values gives more information on the spread of values in the ranges smaller and larger than the median. 

Interquartile Range (IQR) is the difference between the third and first quartile, this gives information on the range of values in the middle half of the data. 

Variance and standard deviation give information on how much of the data is close to the value of the mean. (Standard deviation is equal to the square root of the variance.) 

All of this information can give a numerical picture of the distribution of values in a set of data. (For instance, whether or not the values in the dataset are normally distributed, etc.)

# Question 4

**What are the measures of similarity? Explain the relevance of each of them with suitable examples.**

Distances are used to "measure" similarity. Minkowski distance, Euclidean distance, and Manhattan distance are the measures mentioned in class. 

Euclidean distance is probably the most well-known distance measurement conceptually. It's the same formula for calculating the distance between two points in a graph in common mathematics, except it can be further used for something that has more than 2 dimensions. So the principle of this distance measurement is connecting the two points with a straight line.

The mathematical forumla given by:
$d(i,j) = \sqrt{(x_{i1} - x_{j1})^2 +(x_{i2} - x_{j2})^2+...+(x_{in} - x_{jn})^2}$

Where $i$ and $j$ are $n$-dimensional objects and $d(i,j)$ is the distance between them.

Manhattan distance is also known as the "city block" distance, this is because it gives the distance between two points (in a city) in blocks (rather than a straight line like in Euclidean distance). 

The mathematical formula given by:
$d(i,j) = |x_{i1}-x_{j1}|+|x_{i2}-x_{j2}|+...+|x_{in}-x_{jn}|$

Minkowski distance is the generalization between both the Manhattan distance and the Euclidean distance. 

The mathematical formula given by:
$d(i,j) = (|x_{i1}-x_{j1}|+|x_{i2}-x_{j2}|+...+|x_{in}-x_{jn}|)^{1/p}$

Where $p$ is real number such that $p\ge 1$. 

It's a generalization of the Manhattan and Euclidean distance formulas because you can obtain each formula by substituting $p$ with 1 and 2 respectively. (And apparently $p \to \infty$ can get something called "supremum distance," but this one was not mentioned in class.)
