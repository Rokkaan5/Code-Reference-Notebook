## Question 1

Suppose that the data for analysis includes the attribute temperature. The values are 13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70, 999.

- Use min-max normalization to transform the value 33 for temperature onto the range [0.0,1.0]

**min-max normalization:**
$ v' = \frac{v-min_A}{max_A - min_A}(new\_max_A - new\_min_A)+new\_min_A$

**normalizing 33 to 0-1 scale using min-max:**
$$min_A = 13, max_A=999, new\_max_A = 1, new\_min_A = 0, v=33$$ 

$\Rightarrow v' = \frac{33-13}{ 999- 13}(1.0 - 0)+ 0 $
$\Rightarrow v' = 0.02$

- Use z-score normalization to transform the value 33 temperature

**z-score normalization:**
$ v' = \frac{v - \mu_A}{\sigma_A}$       where $$\mu_A$$ is the mean and $$\sigma_A$$ is the standard deviation

Okay, I'm going to use python to calculate this...

```
import numpy as np

temps = [13,15,16,16,19,20,20,21,22,22,25,25,25,25,30,33,33,35,35,35,35,36,40,45,46,52,70,999]


def z_score(list1, value):
    v = value
    mu = np.mean(list1)
    sigma = np.std(list1)

    return (v-mu)/sigma

print('z-score normalization of 33 is {}'.format(z_score(temps,33)))
```

- Use normalization by decimal scaling to transform the value 33 for temperature

**Decimal scaling:**

$v' = \frac{v}{10^j}$   where j is the smallest integer, such that `max(|v'|)`$<1$


**Normalizing 33 using decimal scaling:**

$\Rightarrow v' = \frac{33}{10^3}$    ($j=3$, because max value (999) has order of magnitude of $10^3$)
$\Rightarrow v'= 0.033$
