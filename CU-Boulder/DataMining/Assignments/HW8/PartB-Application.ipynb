Process the data for the model (hint: the input for the function should be a list, and each item in the list is a transaction, such as `[['a', 'b', 'c'], ['a', 'c', 'd'], ['c', 'd'],...]`)

Admittedly, I found the way to code this from the following link https://www.kaggle.com/code/khusheekapoor/market-basket-analysis-in-python

```python
from csv import reader
```

```python
ts = []
with open ('store_data.csv','r') as f:
    csv_reader = reader(f)
    for row in csv_reader:
        ts.append(row)
```

```python
ts[:4]  # first 5 elements of the 'transaction' list
```

**Conduct apriori for 2-itemset association rules mining (hint: set `max_length=2`), setup your parameters (support, lift, confidence, etc.) so you can find around 10 rules (within [8,12]). Display these rules in a clear format**

```python
rule = apriori(transactions = ts, max_length=2)
result = list(rule)
len(result)
```

According to what I've found, the default parameters for `apriori` are as follows:
- `min_support = 0.1`
- `min_confidence = 0.0`
- `min_lift = 0.0`

So it only makes sense to adjust the `min_support` parameter.

```python
supp = np.round(np.arange(0.01,0.1,0.02),2)

for i in supp:
    rule = apriori(transactions = ts, max_length=2,min_support=i,)
    result = list(rule)
    print('for min_supp = {}, number of rules = {}'.format(i,len(result)))
```

`min_support = 0.09` gives about 10 rules

```python
rule = apriori(transactions = ts, max_length=2,min_support=0.09,)
result = list(rule)
len(result)
```
