# Question 1

**Explain why it is impossible to do FP analysis for all items in Walmart**

The only reason why I can think of why FP analysis on all items in Walmart is "impossible" is because any form of FP analysis requires (at least) one database scan, and scanning a transaction database of all items in Walmart would probably require a lot of computation power (let alone multiple scans). 

(But I'm not completely sure/confident of this answer.)
