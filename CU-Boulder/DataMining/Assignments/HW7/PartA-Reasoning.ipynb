# Question 1

**Identify a real-world business scenario that needs clustering. Assume you are together with the CEO of the business, and you have 2 minutes. Convincing them how your knowledge of clustering can help them.**

Clustering is one of the unsupervised learning methods in machine learning, meaning that it's mainly used for finding information and patterns in the (unlabelled) data. Clustering, in particular, is a way to group certain information together that are similar to each other (within a cluster) but different from other "grouped" information. We can use a clustering model (or several) to group similar products together for some product/app recommendation systems. 

# Question 2

**We learned normalization in data processing. Discuss how scaling impacts some clustering methods, and how it doesn't impact some other clustering methods.**

Clustering models utilizing distance metrics are affected by normalized data. For instance, more specifically, normalization has the potential to improve clustering methods like k-means. From what I understand, density-based clustering and hierarchical clustering are not affected by normalized data.
