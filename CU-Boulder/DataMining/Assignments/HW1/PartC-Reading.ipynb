**Paper title:** *Security and Privacy Issues of Big Data*

# Question 1

**What problem is addressed? Why is the problem important and challenging?**

As suggested by the title, the primary focus of the paper is the challenges in security and privacy that are growing along with the growth of big data. The article brings up the challenges associated with the 5 V's of big data (which also cover the entire data life-cycle) and the issues concerned with the Internet of Things (IoT).

Regarding big data in general, the article points out the categorized challenges outlined by the Cloud Secure Alliance (CSA): infrastructure security, data privacy, data management & integrity, and reactive security. (Which should correspond to the confidentiality, integrity, and availability (CIA) concept in cybersecurity)

Regarding the IoT, the article mentioned the top ten security issues identified with the IoT, known as the "OWASP Internet of Things Top Ten." 

These problems are important because current traditional security infrastructures like firewalls and DMZs are apparently ineffective in these growing and evolving security issues. And they are challenging because they require an infrastructure that can adapt to a heterogenous mix of technology systems and network domains. 


# Question 2

**A high-level, brief description of the proposed solution (no need to include the details).**

The start of the paper mentioned something called "Software Defined Networking" (SDN) to be a solution to the security issues related to big data. Supposedly SDNs allow administrators to write a program that can change the behavior of an entire network from a separate network control logic. This "abstraction" structure (separation structure) of SDNs is supposedly the key point in solving the outlined security issues. (There are more specific details in the "Solutions/Proposals to Address Big Data Security and Privacy Challenges" section of the paper, but since details are not needed for this assignment I will leave it as it is there.)

The paper uses two case studies as ideas/suggestions of specific technologies that utilize this SDN structure that have the potential to contribute to the solutions for the security issues outlined in the paper.
