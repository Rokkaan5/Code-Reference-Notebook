## Question 3

 List out the challenges or difficulties in Data Collection. How would you mine huge data?

I believe one of the challenges in data collection is data integration, which is the process of combining data from multiple data sources. This serves as a challenge because data from different sources can come in different formats and styles (if it's structured). Then there's also the possibility of running into both structured and unstructured data.  

Ways to mine huge (big) data can include things like automated processes using computer algorithms, data visualizations, and analysis of statistical information.
