A student in STEM is looking for an internship/full-time in the same field. You belong to a company that aids students in job search. ### "Tutorial"

Start by importing the necessary libraries

```
import pandas as pd
```

Then load the found datasets. In this case, I found two CSV datasets from Kaggle. One is just labeled "data job posts," and the other is a specific job opening dataset from Indeed supposedly (but had "trulia" in the file name, so I shortened the name to that), both obtained from Kaggle. 

**Just some notes on potentially lacking parts of datasets I found due to not spending an extensive amount of time trying to find the "right" datasets: the datasets are job openings posted several years ago instead of actual recent openings (since they are from Kaggle). They are also all positions with no filtering to types of fields or anything (so not a STEM-specific job opening dataset). 

```
djp_df = pd.read_csv('data job posts.csv')
djp_df
```

```
trulia_df = pd.read_csv('trulia.csv')
trulia_df
```

Pandas in Python allow for the concatenation of data frames, but columns need to have the same name if we want to concatenate columns with similar information to each other. So I'm going to use the `columns` and `info()` methods provided by Pandas to get information on all of the available columns in each dataset and try to compare the types of information given by each dataset. 

```
djp_df.columns
```

```
trulia_df.columns
```

```
djp_df.info()
```

```
trulia_df.info()
```

After some analysis of the available information in the two datasets I found. I decided to keep the overlapping information which was only the job titles, job descriptions, company names, location (which had varying depths of information), and company descriptions. 

Unfortunately, only one of the datasets had any salary information, the other one had (multiple) columns for salary information but contained all NaN values. However, since salary information is still useful I decided to keep the salary information provided by the one dataset (and so when I concatenate the two data frames the salary column for information from the second dataset will just be filled with NaNs). 

```
df1 = djp_df[['Title','JobDescription','Company','Location','AboutC','Salary']]
df2 = trulia_df[['Job Title','Job Description','Location','Company Name','Companydescription']]
```

As mentioned before, if there are columns with similar information but different column names, then the data frames won't concatenate (in a useful way). So, for instance, "Title" and "Job Title" in the data job posts.csv and trulia.csv data respectively refer to the same type of information but don't have the same column name. 

```
df1 = df1.rename(columns = {'Title':'Job Title','JobDescription':'Job Description','Company':'Company Name','AboutC':'Company Description'})
df2 = df2.rename(columns = {'Companydescription': 'Company Description'})
```

```
df1
```

```
df2
```

```
final_df = pd.concat([df1,df2])
final_df
```

Some additional data-cleaning processes might include ways to find ways to filter the information to contain only STEM-related positions. But for time's sake, I'm not going to include that process. (One of the datasets seemed to have a potentially useful column named "industry" that could've helped with that process, but unfortunately, every value was empty.)

One more last look at the "final_df"

```
final_df.info()
```

Finally, use the `to_csv` method from the Pandas library to create a csv file from the finalized dataframe. 

```
final_df.to_csv('jobpostings.csv')
```
