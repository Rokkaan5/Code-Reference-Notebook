# Question 1

**Discuss the big picture of learning and why we have both supervised and unsupervised learning**

Learning, in general, is the process of gaining or extracting information from patterns in past "data" to make some predictions about the future. (A very basic search for a definition of "learning" on google gives the result: "the acquisition of knowledge or skills through experience, study, or by being taught." so if you ultimately consider "experience" as "past data" the concept is just about the same.)

The concepts of supervised and unsupervised learning are based on the dependency on whether the model builder knows the "correct answer" or not. So, supervised learning is training the model using the "correct answers." In other words, the model learns how to make the correct conclusion by finding patterns found when comparing given data with the correct answers on the cheat sheet. 

Unsupervised learning is when the model is not given the correct answer and is pretty much on their own to figure out and pick up patterns it sees in the data it was given. 

As far as why we have two forms of machine learning, one is for making predictions, while the other is gaining information by getting other "perspectives" in finding patterns.

# Question 2

**Someone claims that the "Nearest Neighbor" method is not a learning method. Do you agree or disagree, and why?**

To my understanding, "nearest neighbor" is not considered a learning method by some because it doesn't store any information like patterns or similarities upon getting training data, and thus doesn't even have a "learning phase." And then doesn't do anything until given a test set and only at that point make calculations in comparison to the training data. 

As far as my opinion is concerned, I suppose it's a valid thought. It's hard to think that a model is doing any learning if it doesn't store any information during "training" and so if it doesn't even have a learning stage, then it's difficult to think of it as a learning model. However, if utilized my explanation of learning in the first question, then I would have to consider it learning because the model still utilizes information from "past" data (training data) to make its conclusions.
