**5.17)** Use a computer simulation to verify the proofs for 
$\frac{p}{q} = \frac{1}{3} ; \frac{3}{11}; \sqrt{2}$


```python
import numpy as np
import matplotlib.pyplot as plt

[wx,wy] = [[1,3,np.sqrt(2)],[3,11,1]]
Ax = Ay = 1
d= 0
Ta = 2*np.pi
t = np.arange(0,Ta,1e-5)

def x(Ax,wx,t):
    return Ax*np.cos(wx*t)

def y(Ay,wy,t):
    return Ay*np.cos(wy*t-d)

for i in range(len(wx)-1):
    plt.figure(i)
    plt.plot(x(Ax,wx[i],t),y(Ay,wy[i],t))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('p/q = '+str(wx[i])+'/'+str(wy[i]))
    
```
