**1.50)** The differential equation (1.51) for the skateboard of Example 1.2 cannot be solved in terms of elementary functions, but is easily solved numerically.

---

Equation (1.51): $$\ddot{\phi} = -\frac{g}{R}sin(\phi) $$ 

**(a)** Solve the differential equation for the case that the board is released from $\phi_0=20 ^{\circ}$, using the values $R = 5 m$ and $g = 9.8 m/s^2$. Make a plot of $\phi$ against time for two or three periods.

---

**(b)** On the same picture, plot the approximate solution (1.57) with the same $\phi_0=20 ^{\circ}$. Comment on your two graphs.

Equation (1.57):
$$\phi(t)=\phi_0cos(\omega t)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
#-----------------------------------------------------------------------------------------------------------------------

g=9.8
R=5.

phi0deg=20.                             #phi initial
phi0 = phi0deg/180.*np.pi               #convert phi_0 to radians, b/c python calculates in radians
omg0 = 0                                #omega initial  (intial angular velocity)   

period = 2*np.pi*np.sqrt(R/g)           #Period = 2pi(R/g)^(1/2)

t = np.linspace(0,3*period,10000)       #three periods

def odes(A,t):
    p,w=A
    arr=np.array([w,-g/R*np.sin(p)])
    return arr

#part a
p,w = integrate.odeint(odes,[phi0,omg0],t).T

#part b
omeg = np.sqrt(g/R)
approx = phi0*np.cos(omeg*t)

#graph (both a & b)
#phi converted back to degrees
plt.figure(1,figsize=(15,6))
plt.plot(t,p*180./np.pi,'b-',label='$\phi$ Numerical')                         
plt.plot(t,approx*180./np.pi,'r:',label='$\phi$ Approximate')
plt.title('Problem 1.50 (Skateboard Example Numerical Solution)',fontsize=15)
plt.xlabel('time (sec)',fontsize=12)
plt.ylabel('$\phi$(t) (degrees)',fontsize=12)
plt.legend()
plt.show()
```
