```python
# Homework 1 due Jan. 22, 2020
#-----------------------------------------------------------------------------------------------
import numpy as np  #helps to abbreviate, otherwise the "handle" is numpy.
from matplotlib import pyplot as plt
import math as m
#-----------------------------------------------------------------------------------------------

#-----------------------------------------------------------------------------------------------
def Tn_exp(n,x):  #returns the nth degree Taylor polynomial for the exp function where n=degree.
    i = 0
    f = x**i/m.factorial(i)
    for i in range(1, n):
        f = f + x**i/m.factorial(i)
    return f
#-----------------------------------------------------------------------------------------------

#beginning of main program

n = 6 #degree of Taylor polynomial
x = np.linspace(0,2,100)

fx = Tn_exp(n,x)

#plot results.  Think of plt as an object that encapsulates a group of related procedures or "methods".
plt.plot(x,fx,'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.exp(x),'r')  #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('n degree Taylor polynomial')
plt.show()

```

```python
#graphing for sin(x) Taylor series
import numpy as np  
from matplotlib import pyplot as plt
import math as m

#---------------------------------------------------------------------------------------------------------------------
def Tn_sin(n,x):                    #to return the nth degree Taylor Series for sin(x)
    i=0
    f= ((-1)**i)*(x**(2*i+1))/m.factorial(2*i+1)
    for i in range(1,n):
        f = f + ((-1)**i)*(x**(2*i+1))/m.factorial(2*i+1)
    return f
#---------------------------------------------------------------------------------------------------------------------

n = 8                               #n=degrees
x = np.linspace(0,2*np.pi,100)      #showing graph for x=0 to x=2pi

fx = Tn_sin(n,x)

#plotting the graph-------------------------------------------------------------------------------------------------
plt.plot(x,fx,'b', label= 'Taylor series')                  #blue function is Taylor series representation for sin(x) for specified n
plt.plot(x,np.sin(x),'r', label = 'Actual')           #red function is actual sin(x) over specified domain
plt.xlabel('x')
plt.ylabel('y')
plt.title('n degree Taylor polynomial sin(x)')
plt.legend(fontsize=15)
plt.show()
```

```python
#graphing for cos(x) Taylor series
import numpy as np  
from matplotlib import pyplot as plt
import math as m

#---------------------------------------------------------------------------------------------------------------------
def Tn_cos(n,x):                   #to return the nth degree Taylor Series for cos(x)
    i=0
    f= ((-1)**i)*(x**(2*i))/m.factorial(2*i)
    for i in range(1,n):
        f = f + ((-1)**i)*(x**(2*i))/m.factorial(2*i)
    return f
#---------------------------------------------------------------------------------------------------------------------

n = 8                              #n=degrees
x = np.linspace(0,2*np.pi,100)     #showing graph for x=0 to x=2pi

fx = Tn_cos(n,x)

#plotting the graph-------------------------------------------------------------------------------------------------
plt.plot(x,fx,'b')                 #blue function is Taylor series representation for cos(x) for specified n
plt.plot(x,np.cos(x),'r')          #red function is actual cos(x) over specified domain
plt.xlabel('x')
plt.ylabel('y')
plt.title('n degree Taylor polynomial cos(x)')
plt.show()
```
