# Lab 7: Planetary motion

- The motion of the Sun and Earth is a *two-body problem*.  
- Problem reduced to a one-body problem by: 1) noting the mass of
  the Sun is much greater than the mass of Earth, 2) assume the Sun is
  stationary and fix origin at Sun's center.
- One-body problem is more general.  Introduce reduced mass,

\begin{equation*}
\mu = \frac{mM}{m+M}
\end{equation*}

where m = $6 \times 10^{24}$ kg (Earth) and M = $2 \times 10^{30}$ kg
(Sun). Thus $\mu \approx m$.
- In the following, consider a single particle of mass $m$ moving
  about a fixed center of force, which we take as the origin of the
  coordinate system.
- Newton's universal law of gravitation is

\begin{equation*}
{\rm \bf F} = -\frac{GmM}{r^3} {\rm \bf r} 
\end{equation*}

where ${\rm \bf r}$ is the vector from $M$ to $m$ and $G=6.67 \times
10^{-11}$ m$^3$/(kg s$^2$).
- The force law applies to objects of negligible spatial extent,
  but also applies to any uniform sphere or spherical shell if the
  distance $r$ is measured from the center of each mass.
- Gravitational force has two general properties: 1) its magnitude
  depends only on the separation of the objects and 2) its direction is
  along the line joining the objects.  Such a force is called a
  *central force*. 
- The assumption of central force implies that the orbit (Earth)
  is restricted to a plane (e.g., $xy$) and the *angular momentum*
  is conserved and lies in the third direction ($z$), i.e., 

\begin{equation*}
L_z = ({\rm \bf r} \times m {\rm \bf v})_z = m (x v_y -y v_x)
\end{equation*}

Note that in general $ {\rm \bf L} = {\rm \bf r} \times {\rm \bf p}$ with
${\rm \bf p} = m {\rm \bf v}$.
- Energy conservation applies too, i.e., 

\begin{equation*}
E = \frac{1}{2} m v^2 - \frac{GmM}{r}
\end{equation*}

## Equations of motion

- The equation of motion is

\begin{equation*}
m \frac{d^2 {\rm \bf r}}{dt^2} = - \frac{GmM}{r^3}{\rm \bf r}
\end{equation*}

-  In force in Cartesian coordinates is

\begin{align*}
F_x = - \frac{GmM}{r^2} \cos \theta = -\frac{GmM}{r^3} x \nonumber \\
F_y  =  -\frac{GmM}{r^2} \sin \theta = -\frac{GmM}{r^3} y \nonumber \\
\end{align*}

- Thus the Cartisian equations of motion are

\begin{align*}
\frac{d^2 x}{dt^2} = - \frac{GM}{r^3}x \nonumber \\
\frac{d^2 y}{dt^2} = - \frac{GM}{r^3}y \nonumber 
\end{align*}

## Circular orbits

- Planetary motion is nearly circular, so

\begin{equation*}
a = \frac{v^2}{r}
\end{equation*}

where $v$ is the speed of the object and $a$ is directed towards the
center and is due to the gravitational force, i.e.,

\begin{equation*}
\frac{mv^2}{r} = \frac{GmM}{r^2}
\end{equation*}

or 

\begin{equation*}
v = \left( \frac{GM}{r} \right)^{1/2}
\end{equation*}

- The period of the orbit is

\begin{equation*}
T = \frac{2 \pi r}{v}
\end{equation*}

and 

\begin{equation*}
T^2 = \frac{4 \pi^2 r^3}{GM}
\end{equation*}

-  This is a special case of Kepler's third law where $r$
  corresponds to the semi-major axis of an ellipse.


## Elliptical orbits

- The most general orbit is an ellipse.
- The two foci of an ellipse, $F_1$ and $F_2$, have the property
  that for any point $P$ the distance $F_1P + F_2 P = const$.
- The longer axis is the major axis and half this length is the
  *semi-major axis* $a$.
- The shorter axis is the minor axis and half this length is the
  *semi-minor axis* $b$.
 Specify elliptical orbit by $a$ and the *eccentricity* $e$
  where $e$ is the ratio of the distance between the foci to the
  length of the major axis.
- Since $F_1P + F_2P = 2a$, 

\begin{equation*}
e = \sqrt{1 - \frac{b^2}{a^2}}
\end{equation*}

with $0 < e < 1$.  For a circle $e=0$.  For Earth's orbit, $e = 0.0167$.

## Astronomical Units

- Since large numbers (once upon a time) are awkward, it is
  convenient to choose a system of units in which the magnitude of the
  produce $GM$ is the order of unity.
- Convenient unit of length is the *astronomical unit* with

\begin{equation*}
1 {~\rm AU} = 1.496 \times 10^{11} {~{\rm m}}
\end{equation*} 

-  The unit of time is one year = $3.15 \times 10^7$ s.  
-  In these units, $T$ = one year, $a$ = 1 AU, 

\begin{equation*}
GM = \frac{4 \pi^2 a^3}{T^2} = 4 \pi^2 {\rm AU^3/yr^2}
\end{equation*}

## Mini solar system

- The presence of other planets implies that the total force on a
  planet is no longer a central force.
- Furthermore, the orbits of the planets are not exactly in the
  same plane, so the analysis must be extended to three dimensions.
- For simplicity, consider two planets of mass $m_1$ and $m_2$.
  The equations of motion are

\begin{align*}
m_1 \frac{d^2 {\rm \bf r}_1}{dt^2} = - \frac{m_1G  M}{r_1^3} {\rm \bf r}_1 +
  \frac{m_1Gm_2}{r_{21}^3} {\rm \bf r}_{21} \nonumber \\
m_2 \frac{d^2 {\rm \bf r}_2}{dt^2} = - \frac{m_2G  M}{r_2^3} {\rm \bf r}_2 -
  \frac{m_2Gm_1}{r_{21}^3} {\rm \bf r}_{21} \nonumber \\
\end{align*}

where ${\rm \bf r}_1$ and ${\rm \bf r}_2$ are directed from the Sun to the
planets and ${\rm \bf r}_{21} = {\rm \bf r}_2 - {\rm \bf r}_1$.
- No analytical solution exists!  Requires a numerical solution.


## Introduction to Python "classes".

````python
#Name: Jasmine Kobayashi
#Date: 2/27/20
#======================================================================================================================
import numpy as np
import matplotlib.pyplot as plt

class planet:
    
    """
    define a class planet that contains initial position and velocity as well as mass and color.
    The class is called with arguments (x0,y0,vx0,vy0,m,color)
    """
    GM = 4*np.pi**2
    
    def __init__(self,x0,y0,vx0,vy0,m,clr):
        self.xn = x0
        self.yn = y0
        self.vxn = vx0
        self.vyn = vy0
        self.m = m
        self.clr = clr
        self.x = []
        self.y = []
        self.vx = []
        self.vy = []
        self.time = []
        

    def F(self,r,x,v):
        return np.array([v,-GM*x/r**3])
    
    def rk4_step(self,r,x,v):           #Just the rk4 function
        k1 = dt*self.F(r,x,v)         
        k2 = dt*self.F(r,x + 0.5*k1[0], v + 0.5*k1[1])
        k3 = dt*self.F(r,x + 0.5*k2[0], v + 0.5*k2[1])
        k4 = dt*self.F(r,x + k3[0], v + k3[1])
        x = x + (k1[0] + 2*k2[0] + 2*k3[0] + k4[0])/6.
        v = v + (k1[1] + 2*k2[1] + 2*k3[1] + k4[1])/6.
        return x,v
        
    def rk4_planet(self,tfinal,dt):     #updating position and velocity arrays until tfinal using rk4 function
        t = 0.0
        while t < tfinal+dt:
            self.time.append(t)
            self.x.append(self.xn)
            self.y.append(self.yn)
            self.vx.append(self.vxn)
            self.vy.append(self.vyn)
            r = np.sqrt(self.xn**2 + self.yn**2)
            self.xn,self.vxn = self.rk4_step(r,self.xn,self.vxn)  #x-component
            self.yn,self.vyn = self.rk4_step(r,self.yn,self.vyn)  #y-component
            t = t + dt

    def plot_planet(self,nstep):
        fig = plt.figure()
        ax = fig.add_subplot(111)
        ax.plot(self.x[::nstep],self.y[::nstep],'.',markersize=0.5,color=self.clr)
        ax.set_xlabel('x (AU)')
        ax.set_ylabel('y (AU)')
        ax.set_aspect('equal')
        ax.plot([0,0],[0,0],'o',markersize=10.0,color='gold')
        ax.minorticks_on()
        ax.grid(True, which='both')
        plt.show()
        
        
#Exercise 1; part 3: Conservation of Energy method
    def energy_con(self):
        self.ve = []             #total velocity array
        self.rs = []             #radial distance from the sun
        self.E = []              #total energy 
        self.dif = []            #difference in energy
        for i in range(0,len(self.vx)):
            self.rs.append(np.sqrt(self.x[i]**2 + self.y[i]**2))
            self.ve.append(np.sqrt((self.vx[i])**2 +(self.vy[i])**2))        #v = (vx^2 + vy^2)^(1/2)
            self.E0 = (0.5*self.m*((self.ve[0])**2)) - (GM*self.m/self.rs[0])         #E0 = 0.5mv0^2 - GmM/r
            self.E.append((0.5*self.m*((self.ve[i])**2)) - (GM*self.m/self.rs[i]))
            self.dif.append(abs((self.E[i]-self.E0)/self.E0))

    def plot_energy_c(self):     #plotting change in total energy
        plt.plot(self.time,self.dif)
        plt.title("Energy conservation")
        plt.xlabel("time")
        plt.ylabel("$\Delta^n$")
        
#Exercise 1; part 4: Conservation of Angular momentum
    def ang_moment(self):
        self.L = []             #Angular momentum
        self.L_diff = []        #change in angular momentum
        for j in range(0,len(self.vx)):
            self.L0 = self.m*((self.x[0]*self.vy[0]) - (self.y[0]*self.vx[0]))
            self.L.append(self.m*((self.x[j]*self.vy[j]) - (self.y[j]*self.vx[j])))
            self.L_diff.append(abs((self.L[j]-self.L0)/self.L0))

    def plot_am(self):          #plotting difference in angular momentum
        plt.plot(self.time,self.L_diff)
        plt.title("Difference in Angular momentum vs. time")
        plt.xlabel("time")
        plt.ylabel("$\Delta$ L")
        
#Exercise 1; part 5: Verifying Kepler's second law (equal areas in equal times)
    def kepler_2nd(self,):
        a = int(len(self.rs)/500)                #amount of time passed
        A1 = round(0.5*self.rs[a]*self.ve[a],6)  #dA/dt = (1/2)r(r(dtheta/dt)) = (1/2)r(v)
        A2 = round(0.5*self.L[a]/self.m,6)       #dA/dt = (1/2)(L/m)    #L = ang. momentum; m = mass
        print("For verifying Kepler's Second law: Area 1 =",A1,"Area 2=",A2)

#Exercise 1; part 6: Kepler's third law (period^2 proportional to (semimajor axis)^3)
    def kepler_3rd(self,):
        self.所diff = []
        for k in range(0,len(self.x)):
            self.所diff.append(np.sqrt((self.x[k]-self.x[0])**2 + (self.y[k]-self.y[0])**2))
        self.major = max(self.所diff)      #major axis of elliptical orbit = max difference between all positions and initial position
        self.smaj = self.major/2           #semi major axis
        #print(self.smaj)
        self.T = 2*np.pi*self.rs[0]/self.ve[0]     #T = 2pi*r/v
        self.k = (self.T**2)/(self.smaj**3)        #since proportional: T^2 = k(a^3), so k = T^2/a^3
        print("For Kepler 3rd law: Period(T) squared=",self.T**2,"and semi-major axis(a) cubed =",self.smaj**3)
        print("T^2/a^3=",self.k)
        
````
