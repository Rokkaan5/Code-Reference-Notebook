### Ch.7: 3.3

For following combination of a fundamental musical tone and some of its overtones, make a computer plot of individual harmonics (all on the same axes) and then a plot of the sum. Note that the sum has the period of the fundamental.

**3.3)** $\sin(\pi t) + \sin(2 \pi t) + \frac{1}{3} \sin(3 \pi t)$

```python
import numpy as np
import matplotlib.pyplot as plt

#Range of graph
t = np.arange(0,2*np.pi,0.001)

#Individual harmonics
harmonic1 = np.sin(np.pi*t)
harmonic2 = np.sin(2*np.pi*t)
harmonic3 = (1/3)*np.sin(2*np.pi*t)

#Sum of harmonics
combined = harmonic1 + harmonic2 + harmonic3

#Graphing
plt.figure(1,figsize=(15,6))
plt.plot(t,harmonic1,"--",label = "$sin(\pi t)$")
plt.plot(t,harmonic2,"--",label = "$sin(2\pi t)$")
plt.plot(t,harmonic3,"--",label = "$(1/3)sin(3\pi t)$")
plt.plot(t,combined,label = "sum")
plt.legend(fontsize = 15)
plt.title("Ch.7: 3.3",fontsize = 15)
plt.xlabel("t",fontsize=12)
plt.ylabel("Harmonic",fontsize=12)
plt.grid(True)
```
