```python
import numpy as np
import matplotlib.pyplot as plt

#Vialov Data
dist = [50,100,150,200,250,350,450,550,700,850,950]
thickness = [0.8,1.5,1.8,2.1,2.3,2.6,2.9,3.1,3.4,3.5,3.5]

L = 1000       #(km)
H = 3500e-3    #(km)

n = 3
x = np.arange(50,L,1)  #(km)

def h(x):
    first = (H/np.sqrt(L))
    inex = (n+1)/n
    oex = n/((2*n)+2)
    second = ((L**inex)-(L-x)**inex)**oex
    height = first*second
    return height

def ideal(x):
    tau = 1e5
    g = 9.81e3   #(km/s^2)?
    rho = 900    #(kg/m^3)?
    return np.sqrt((2*tau/(rho*g))*x)

vialov = []
plastic = []
for i in range(len(x)):
    vialov.append(h(x[i]))
    plastic.append(ideal(x[i]))


plt.figure()
plt.plot(dist,thickness,"o",label="Vialov Data")
plt.plot(x,vialov,label="Vialov profile")
plt.plot(x,plastic,label="Ideal Plastic")
plt.title("Comparison graph with a real ice sheet profile")
plt.xlabel("Distance (km)")
plt.ylabel("Thickness (km)")
plt.legend()
```
