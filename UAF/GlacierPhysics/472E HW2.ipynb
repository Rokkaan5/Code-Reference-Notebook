```python
import numpy as np
import matplotlib.pyplot as plt

#Vialov Data
dist = [50,100,150,200,250,350,450,550,700,850,950]
thickness = [0.8,1.5,1.8,2.1,2.3,2.6,2.9,3.1,3.4,3.5,3.5]

L = 1000       #(km)
H = 3500e-3    #(km)

n = 3
g = 9.81
rho = 900
x = np.arange(50,L,1)

def h(x):
    exp = n/((2*n)+2)
    inex = (n+1)/n
    first = (2/(rho*g))**exp
    second = ((L**inex)-((L-x)**inex))**exp
    third = ((n+1)/(1-(1/(n+2))))**(1/(2*n+2))
    height = first*second
    return height

plastic = []
for i in range(len(x)):
    plastic.append(h(x[i]))

plt.figure()
plt.plot(dist,thickness,"o",label="Vialov Data")
plt.plot(x,plastic,label="ideal plastic")
plt.title("Comparison graph with a real ice sheet profile")
plt.xlabel("Distance (km)")
plt.ylabel("Thickness (km)")
plt.legend()
```
